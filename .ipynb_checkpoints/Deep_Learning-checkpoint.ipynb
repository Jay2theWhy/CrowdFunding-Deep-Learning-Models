**Defining Some Functions**

get_wordnet_pos: obtain the tag value given a word

auc_roc: create an ROC curve and calculate the area under the curve (AUC)

plot_history: plot the training and validation loss of the model

heat_map: create a confusion matrix

```python
def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)
  
def auc_roc(model, X_test, X_train, y_train, y_test):
    # Calculate outputs in test set
    prob_test = model.predict(X_test, verbose = 1)
    prob_train = model.predict(X_train, verbose = 1)

    # Compute ROC curve and ROC area
    fpr, tpr, _ = roc_curve(y_train, prob_train)
    roc_auc = auc(fpr, tpr)
    print('\nAUC train: ', roc_auc)

    # Compute ROC curve and ROC area
    fpr, tpr, _ = roc_curve(y_test, prob_test)
    roc_auc = auc(fpr, tpr)
    print('AUC test: ', roc_auc)
    return roc_auc, fpr, tpr

def plot_history(model):
    # Plot history

    print(model.history.history.keys())

    sns.set_style("darkgrid")
    plt.plot(model.history.history['loss'], label='Training Loss')

    sns.set_style("darkgrid")
    plt.plot(model.history.history['val_loss'], label='Validation Loss')
    plt.legend()
    plt.title('Training and Validation Loss')

    plt.show()

def heat_map(y, y_pred, axis):
    # Calculate confusion matrix
    confusion_m = confusion_matrix(y, 
                        y_pred)

    # Turn matrix to percentages
    confusion_m = confusion_m.astype('float') / confusion_m.sum(axis=1)[:, np.newaxis]

    # Turn to dataframe
    df_cm = pd.DataFrame(
            confusion_m, index=axis, columns=axis, 
    )

    # Parameters of the image
    figsize = (10,7)
    fontsize=14

    # Create image
    fig = plt.figure(figsize=figsize)
    heatmap = sns.heatmap(df_cm, annot=True, fmt='.2f')

    # Make it nicer
    heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, 
                                ha='right', fontsize=fontsize)
    heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=45,
                                ha='right', fontsize=fontsize)

    # Add labels
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

    # Plot!
    plt.show()
```

**Data Preprocessing**
1. Use Regular Expressions to remove website links, usernames, and numbers
2. Remove all punctuation characters from the text, remove any double spaces, and change all characters to lowercase
3. Tokenize the sentences into individual words so each word can be processed
4. Lemmatization performed on all words to reduce them to their root words
5. Remove stop words
6. Untokenize all words back into whole sentences

```python
###---Data Preprocessing---###

## Remove things that aren't meant to be words using RegEx
df['text'] = [re.sub('@[a-zA-Z0-9]+[\s]?[A-Z]?[a-z]*', '', x) for x in df['text']] # removing username tags
df['text'] = [re.sub('(https:)[^\s]+', '', x) for x in df['text']] # removing website links starting with https:
df['text'] = [re.sub('(www.)[^\s]+', '', x) for x in df['text']] # removing website links starting with www.
df['text'] = [re.sub('[0-9]+', '', x) for x in df['text']] # remove all numbers

## remove punctations
# Collect punctuation signs.
table = str.maketrans(' ', ' ', string.punctuation)
# Remove them from the text
df.iloc[:,0] = [j.translate(table) for j in df.iloc[:,0]]
df.iloc[:,0] = [j.replace('\x96',' ') for j in df.iloc[:,0]]
# Eliminate double spaces
df.iloc[:,0] = [" ".join(j.split()) for j in df.iloc[:,0]]
# change everything to lowercase
df['text'] = [str.lower(x) for x in df['text']]

## Tokenization (splitting sentences into individual words)
df['words'] = [word_tokenize(x) for x in df.iloc[:,0]]

## Removing Stop Words
stop_words = set(stopwords.words('english'))
df['words'] = [[y for y in x if not y in stop_words] for x in df['words']]

## Lemmatization (converting words to their root words)
lemmatizer = WordNetLemmatizer() 
df['words'] = [[lemmatizer.lemmatize(y, get_wordnet_pos(y)) for y in x] for x in df['words']]

## Untokenize all words back into sentences
df['processed_text'] = [TreebankWordDetokenizer().detokenize(x) for x in df['words']]
```

**Neural Network 1: Sequential Convolutional Layers**

Rectified Linear Unit (ReLU) activation function was used initially due to its efficiency and speed. The outputs were then fed into a sigmoid activation function since the response variable (Sentiment) is binary. 

Model utilized three convolutional layers with varying kernel sizes sandwiched between two dense layers.


```python
def NN_Model_1(embedding_matrix):
    model = Sequential()
    embedding_layer = Embedding(len(embedding_matrix),      
                                300,                           
                                weights=[embedding_matrix],    
                                input_length=100,              
                                trainable=False)               

    model.add(embedding_layer)

    model.add(Dense(256, activation = 'relu'))

    model.add(Conv1D(filters=256,          
                    kernel_size=3,         
                    padding='valid',       
                    activation='relu',     
                    strides=1)             
            )

    model.add(Conv1D(filters=256,          
                    kernel_size=2,         
                    padding='valid',       
                    activation='relu',     
                    strides=1)             
            )

    model.add(Conv1D(filters=64,           
                    kernel_size=2,         
                    padding='valid',       
                    activation='relu',     
                    strides=1)             
            )

    model.add(MaxPooling1D(pool_size = 3))

    # Flatten
    model.add(Flatten())

    # Dense Layer of size 256 with Dropout
    model.add(Dense(256, activation = 'relu'))
    model.add(Dropout(0.5))

    # Output layer of size 1
    model.add(Dense(1, activation = 'sigmoid'))

    model.compile(loss='binary_crossentropy',
                  optimizer='adam',
                  metrics=['acc'])
    return model
```

**Neural Network 2: Parallel Convolutional Layers**

ReLU and sigmoid activation functions were used again for the same reasons as the previous model.

This model utilized a parallel layer containing three separate streams that each contained with two convolutional layers with different parameters. Each stream had a different kernel size, thus resulting in the layer looking at a different number of words at a time. This parallel layer was sandwiched between two dense layers, in a similar fashion as the previous model.

```python
### Parallel Model ###
# Filter sizes to use.
filter_sizes = (2,3,5)

# Initialize
graph_in = Input(shape=(100, 256))
convs = []
avgs = []

for fsz in filter_sizes:
    conv1 = Conv1D(filters=256,
                   kernel_size=fsz,
                   padding='valid',
                   activation='relu',
                   strides=1)(graph_in) 
    conv2 = Conv1D(filters=128,
                   kernel_size=3,
                   padding='valid',
                   activation='relu',
                   strides=1)(conv1)
    pool = MaxPooling1D(pool_size=fsz)(conv2) 
    flattenMax = Flatten()(pool) 
    convs.append(flattenMax) 

# Concatenate layers.
if len(filter_sizes)>1:
    out = Concatenate()(convs)
else:
    out = convs[0]

graph = Model(inputs=graph_in, outputs=out, name="graphModel")

def NN_Model_2(embedding_matrix):
    model = Sequential()
    embedding_layer = Embedding(len(embedding_matrix),         
                                300,                           
                                weights=[embedding_matrix],    
                                input_length=100,              
                                trainable=False)               

    model.add(embedding_layer)

    model.add(Dense(256, activation = 'relu'))

    model.add(graph)

    model.add(Dense(256, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(loss='binary_crossentropy',
                  optimizer='adam',
                  metrics=['acc'])
    
    return model
```
